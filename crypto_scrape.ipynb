```python
from bs4 import BeautifulSoup
import requests
print(requests.__version__)
from datetime import datetime
```

```python
url='https://coinmarketcap.com/'
page=requests.get(url)
```

```python
cryptoSoup=BeautifulSoup(page.content, 'html.parser')
print(cryptoSoup.prettify())
```

```python
priceBox=cryptoSoup.find("div", class_="sc-4c05d6ef-0 bLqliP")
print((priceBox.prettify()))
```

```python
# Coin Name Full
coinName=cryptoSoup.find_all("p",class_="sc-71024e3e-0 ehyBa-d")

print(coinName[:10])
```

```python
for x,y in enumerate(coinName):
    temp=coinName[x].text
    # print(temp)
    coinName[x]=temp
print(coinName)
```

```python
# Coin Name Short
nameShort=cryptoSoup.find_all("p", class_="sc-71024e3e-0 OqPKt coin-item-symbol")

print(nameShort[:10])
```

```python
for x,y in enumerate(nameShort):
    temp=nameShort[x].text
    nameShort[x]=temp
print(nameShort)
```

```python
# Coin Price Box
coinBox=cryptoSoup.find_all("div", class_="sc-a093f09c-0 gPTgRa")

print(coinBox[:10])
```

```python
for x,y in enumerate(coinBox):
    temp=coinBox[x].text
    # print(temp)
    coinBox[x]=temp
print(coinBox)
```

```python
coinNameConcat=[]
for i,value in enumerate(coinName):
    coinNameConcat.append(value+" ("+nameShort[i]+")")

print(coinNameConcat)
```

```python
coinName_priceList={}

for key,value in zip(coinNameConcat,coinBox):
    coinName_priceList[key]=(value)
print(coinName_priceList)
```

```python
print(coinName_priceList)
```

```python
#Date time log item
coinName_priceList["DATE_TIME"]=datetime.now()
```
